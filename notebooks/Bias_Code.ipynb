```python
first_time_bloc = ['08:00', '08:15', '08:30', '08:45', '09:00', '09:15', '09:30', '09:45', '10:00', '10:15', '10:30', '10:45', '11:00', '11:15', '11:30', '11:45', '12:00', '12:15', '12:30', '12:45', '13:00', '13:15', '13:30', '13:45', '14:00', '14:15', '14:30', '14:45', '15:00', '15:15', '15:30', '15:45']
second_time_bloc = ['00:00', '00:15', '00:30', '00:45', '01:00', '01:15', '01:30', '01:45', '02:00', '02:15', '02:30', '02:45', '03:00', '03:15', '03:30', '03:45', '04:00', '04:15', '04:30', '04:45', '05:00', '05:15', '05:30', '05:45', '06:00', '06:15', '06:30', '06:45', '07:00', '07:15', '07:30', '07:45']
third_time_bloc = ['16:00', '16:15', '16:30', '16:45', '17:00', '17:15', '17:30', '17:45', '18:00', '18:15', '18:30', '18:45']
fourth_time_bloc = ['22:00', '22:15', '22:30', '22:45', '23:00', '23:15', '23:30', '23:45']
```

```python
def calc_bias(row):
    total_bias = 0
    
    rating_bias = -0.7 if row['Rating'] == 1 else -0.5 if row['Rating'] == 2 else -0.2 if row['Rating'] == 3 else 0.3 if row['Rating'] == 4 else 0.8
    
    price_paid_bias = row['Cost of Purchase'] * rating_bias * 0.01 # high price_paid could be a satisfier or dissatisfier based on rating
    
    time_of_transaction_bias = -0.25 if row["Time of Transaction"] in first_time_bloc else -0.16 if row["Time of Transaction"] in second_time_bloc else -0.04 if row["Time of Transaction"] in third_time_bloc else 0.05 if row["Time of Transaction"] in fourth_time_bloc else 0.1
    
    time_taken_bias = (0.5964 - row['Time Taken for Transaction (s)'] * 0.01) * time_of_transaction_bias ## time taken bias is affected by when the transaction took place
    
    total_bias = time_of_transaction_bias + rating_bias + price_paid_bias + time_taken_bias
    
    return total_bias
```

```python
df['bias'] = df.apply(calc_bias, axis = 1)
df.head()
```

```python
labels = []

for idx, row in df.iterrows():
    
    if 0.5 + row['bias'] > 1: #confirmation of certainty when probability > 1
        a = 1
        
    elif 0.5 + row['bias'] < 0: #prevention of negative probabilities
        a = 0
        
    else:
        a = 0.5 + row['bias']
        
    randomness = (0.4 * np.random.random_sample() - 0.2) #addition of randomness
    a += randomness
    
    #b = 1 - a
    #labels.append(np.random.choice([1, 0], p = [a, b])) if using probability, model will work exactly as well as the number of positive classes
    labels.append(1 if a > 0.5 else 0)
    

df['return'] = labels
df.drop('bias', axis = 1, inplace = True)
```
